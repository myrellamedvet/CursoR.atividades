```r
# Definir o caminho da pasta do projeto
caminho_projeto <- "C:/meu_projeto"

# Verificar se a pasta existe
if (!dir.exists(caminho_projeto)) {

  # Se NÃO existir → cria a pasta
  dir.create(caminho_projeto, recursive = TRUE)

  cat("Pasta criada com sucesso!\n")

} else {

  cat("A pasta já existe.\n")
}

# Configurar o diretório de trabalho para a pasta correta
setwd(caminho_projeto)

# Conferir para qual pasta o projeto está apontando
getwd()
```
