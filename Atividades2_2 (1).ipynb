```r
# ---------------------------
# CONFIG
# ---------------------------
required_packages <- c("dplyr", "ggplot2")

# ---------------------------
# READ → verifica se está instalado
# ---------------------------
pkg_instalado <- function(pkg) {
  pkg %in% rownames(installed.packages())
}

# ---------------------------
# CREATE → instala pacote
# ---------------------------
instalar_pkg <- function(pkg) {
  install.packages(pkg, dependencies = TRUE)
  message(paste("Pacote", pkg, "instalado"))
}

# ---------------------------
# LOAD → carrega pacote
# ---------------------------
carregar_pkg <- function(pkg) {
  library(pkg, character.only = TRUE)
  message(paste("Pacote", pkg, "carregado"))
}

# ---------------------------
# UPDATE → reinstala (opcional)
# ---------------------------
atualizar_pkg <- function(pkg) {
  install.packages(pkg)
  message(paste("Pacote", pkg, "atualizado"))
}

# ---------------------------
# CONTROLE (orquestração)
# ---------------------------
gerenciar_pkg <- function(pkg) {

  if (!pkg_instalado(pkg)) {

    instalar_pkg(pkg)     # CREATE
    carregar_pkg(pkg)     # LOAD

  } else {

    message(paste("Pacote", pkg, "já instalado"))
    carregar_pkg(pkg)     # READ + LOAD

  }
}

# ---------------------------
# EXECUÇÃO (linha por linha)
# ---------------------------
for (pkg in required_packages) {
  gerenciar_pkg(pkg)
}
```
